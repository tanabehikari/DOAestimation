```python
import numpy as np
from libs.utils import write__float32_as_int16,read_int16_as_float32, write__float32_as_int16
%matplotlib inline
import matplotlib.pyplot as plt
import IPython.display
```

```python
x_org=read_int16_as_float32('nf144032.wav')

x1=read_int16_as_float32('mvdr/10dB/nf144032.1.wav')
x2=read_int16_as_float32('mvdr/m10dB/nf144032.1.wav')
```

```python
plt.subplot(3,1,1)
plt.plot(x_org)
plt.subplot(3,1,2)
plt.plot(x1)
plt.subplot(3,1,3)
plt.plot(x2)
IPython.display.Audio(x_org,rate=16000)
```

```python
norm_1 = np.linalg.norm(x1)
norm_2 = np.linalg.norm(x2)
norm_org = np.linalg.norm(x_org)
print("norm of 10dB.wav: {}".format(norm_1))
print("norm of -10dB.wav: {}".format(norm_2))
print("norm of org.wav: {}".format(norm_org))

print("\n After Normalized")
normalized_norm_1 = np.linalg.norm(x1/norm_1*norm_org)
normalized_norm_2 = np.linalg.norm(x2/norm_2*norm_org)
print("norm of 10dB.wav: {}".format(normalized_norm_1))
print("norm of -10dB.wav: {}".format(normalized_norm_2))
```

```python
plt.subplot(3,1,1)
plt.plot(x_org)
plt.subplot(3,1,2)
plt.plot(x1 / norm_1 * norm_org)
plt.subplot(3,1,3)
plt.plot(x2/ norm_2 * norm_org)
```

```python
write__float32_as_int16('normalized_x1.wav', 16000, x1/norm_1*norm_org)
write__float32_as_int16('normalized_x2.wav', 16000, x2/norm_2*norm_org)
```

```python
IPython.display.Audio(x1/norm_1*norm_org,rate=16000)
```
